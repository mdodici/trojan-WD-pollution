```python
import rebound
import reboundx
import numpy as np

from IPython.display import display, clear_output
import matplotlib.pyplot as plt
%matplotlib inline
```

## Check J2 is working properly

```python
simJ2 = rebound.Simulation()

simJ2.add(m=9.543e-4, x=0, y=0, z=0, vx=0, vy=0, vz=0, hash='jupiter')
simJ2.add(m=3e-16, primary=simJ2.particles[0], a=0.05, e=.6, inc = .3, Omega = .6, omega = .2, hash='moon')
simJ2.dt = 0.06
simJ2.ri_ias15.epsilon=0

psJ2 = simJ2.particles
```

```python
rebx = reboundx.Extras(simJ2)

gh = rebx.load_force("gravitational_harmonics")
rebx.add_force(gh)
```

```python
psJ2['jupiter'].params["J2"] = 14736e-6
psJ2['jupiter'].params["R_eq"] = 0.000477895
```

```python
Nout = 1000
timesJ = np.linspace(0., 1e4, Nout)
pomegasJ = np.zeros(Nout)
EerrJ = np.zeros(Nout)
E0J = simJ2.calculate_energy() + rebx.gravitational_harmonics_potential()

for i, time in enumerate(timesJ):
    simJ2.integrate(time)
    EfJ = simJ2.calculate_energy() + rebx.gravitational_harmonics_potential()
    pomegasJ[i] = psJ2["moon"].pomega
    EerrJ[i] = abs(EfJ-E0J)/abs(E0J)
    
simJ2.status()
```

```python
predrate = psJ2["moon"].n*(1.5*psJ2["jupiter"].params["J2"]*(psJ2["jupiter"].params["R_eq"]/(psJ2["moon"].a*(1-psJ2["moon"].e**2)))**2*np.cos(psJ2["moon"].inc))
# predrate from equation shown here
# https://ai-solutions.com/_freeflyeruniversityguide/j2_perturbation.htm
predpomegaJ = [predrate*time+pomegasJ[0] for time in timesJ]

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(timesJ, pomegasJ, '.', label='Integration')
ax.plot(timesJ, predpomegaJ, linewidth=5,  label='Prediction')
ax.set_xlabel('Time', fontsize=24)
ax.set_ylabel('Longitude of Pericenter', fontsize=24)
ax.legend(fontsize=24)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='polar')
c = ax.plot(pomegasJ, timesJ)
p = ax.plot(predpomegaJ, timesJ)
```

## Check that the Sun is working properly

```python
simSun = rebound.Simulation()

simSun.add(m=9.543e-4, x=0, y=0, z=0, vx=0, vy=0, vz=0, hash='jupiter')
simSun.add(m=0, primary=simSun.particles[0], a=0.05, e=.6, inc = .3, Omega = .6, omega = .2, hash='moon')
simSun.add(m=1,primary=simSun.particles[0], a=5.2, e=.04839, inc=.022689, Omega=1.755, omega=.2574, hash='Sun')
simSun.dt = 0.06
simSun.ri_ias15.epsilon=0

psSun = simSun.particles

sunm0 = psSun["Sun"].m
jupm0 = psSun["jupiter"].m
a_moon0 = psSun["moon"].a
a_sun0 = psSun["Sun"].a
n_moon0 = psSun["moon"].n
```

```python
Nout = 1000
timesS = np.linspace(0., 1e4, Nout)
pomegasS = np.zeros(Nout)

for i, time in enumerate(timesS):
    simSun.integrate(time)
    pomegasS[i] = psSun["moon"].pomega

simSun.status()
```

```python
predrate = psSun["moon"].n*(.75*(psSun["Sun"].m/psSun["jupiter"].m)*(psSun["moon"].a/psSun["Sun"].a)**3)
predpomega = [predrate*time for time in timesS]

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(timesS, np.sin(pomegasS), '.', label='Integration')
ax.plot(timesS, np.sin(predpomega), linewidth=5,  label='Prediction')
ax.set_xlabel('Time', fontsize=24)
ax.set_ylabel('Longitude of Pericenter', fontsize=24)
ax.legend(fontsize=24)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='polar')
c = ax.plot(pomegasS, timesS)
p = ax.plot(predpomega, timesS)
ax.grid(False)
```

## Check that mass loss is happening, and is right

```python
sim = rebound.Simulation()

sim.add(m=9.543e-4, x=0, y=0, z=0, vx=0, vy=0, vz=0, hash='jupiter')
sim.add(m=0, primary=sim.particles[0], a=0.086565, e=.16, inc = 175.2*(np.pi/180), Omega = .6, omega = .2, hash='moon1')
sim.add(m=1,primary=sim.particles[0], a=5.2, e=.04839, inc=.022689, Omega=1.755, omega=.2574, hash='Sun')
sim.dt = 0.06
sim.ri_ias15.epsilon=0
M0 = sim.particles[2].m

ps = sim.particles
```

```python
rebx = reboundx.Extras(sim)

modifymass = rebx.load_operator("modify_mass")
rebx.add_operator(modifymass)
```

```python
ps['Sun'].params["tau_mass"] = -1e4
```

```python
Nout = 1000
mass = np.zeros(Nout)
a_moon = np.zeros(Nout)
a_sun = np.zeros(Nout)
e_moon = np.zeros(Nout)
i_moon = np.zeros(Nout)
times = np.linspace(0., 1e4, Nout)

for i, time in enumerate(times):
    sim.integrate(time)
    mass[i] = sim.particles[2].m
    a_sun[i] = sim.particles[2].a
    a_moon[i] = sim.particles[1].a
    e_moon[i] = sim.particles[1].e
    i_moon[i] = sim.particles[1].inc
    
sim.status()
```

```python
pred = M0*np.e**(times/ps[2].params["tau_mass"])

fig, ax = plt.subplots(5,1,figsize=(15,20), sharex=True)

ax[0].plot(times,mass, label='simulation')
ax[0].plot(times,pred, label='predicted')
ax[0].set_ylabel("Star's Mass (MSun)", fontsize=16)
ax[1].plot(times,a_sun, label='simulation')
ax[1].set_ylabel(r"$a_{Jup,Sun}$" + " (AU)", fontsize=16)
ax[2].plot(times,a_moon, label='simulation')
ax[2].set_ylabel(r"$a_{Moon,Jup}$" + " (AU)", fontsize=16)
ax[3].plot(times,e_moon, label='simulation')
ax[3].set_ylabel(r"$ecc_{moon}$", fontsize=16)
ax[4].plot(times,i_moon, label='simulation')
ax[4].set_ylabel(r"$inc_{moon}$", fontsize=16)
ax[4].set_xlabel("Time (yrs)", fontsize=16)

plt.tight_layout()
```
