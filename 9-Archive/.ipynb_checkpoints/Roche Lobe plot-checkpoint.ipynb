```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
%matplotlib inline
```

Using equation from this pset: http://www.physics.usyd.edu.au/~helenj/SeniorAstro/lab4.pdf

```python
G = 4*np.pi**2

def roche(x,y,M1,M2,a):
    q = M2/M1
    xc = q*a/(1+q)
    s1 = np.sqrt(x**2 + y**2)
    s2 = np.sqrt((x-a)**2 + y**2)
    
    term1 = G*(M1/s1 + M2/s2)
    term2 = G*M1/(2*a**3)*((x-xc)**2 + y**2)
    return -term1 - term2

def roch2(x,y,M1,M2,a):
    q = M2/M1
    x /= a
    y /= a
    s1 = np.sqrt(x**2 + y**2)
    s2 = np.sqrt((x-1)**2 + y**2)
    
    term1 = 2/(s1*(1+q))
    term2 = 2*q/(s2*(1+q))
    term3 = (x - q/(1+q))**2
    term4 = y**2
    return term1 + term2 + term3 + term4

def droch(x,y,M1,M2,a):
    q = M2/M1
    s1 = x**2 + y**2
    s2 = (x-a)**2 + y**2
    xc = a*q/(1+q)
    
    term1 = G*(-2*M1*x/(s1**2) - 2*M2*(x-a)/(s2**2))
    term2 = (G*M1/(a**3))*(x - x*a*q/(1+q))
    return -term1 - term2

def dpot2(x,y,M1,M2,a):
    q = M2/M1
    x /= a
    y /= a
    s1 = x**2 + y**2
    s2 = (x-1)**2 + y**2
    
    term1 = (2/(1+q))*(-2*x/(s1**2) - 2*q*(x-1)/(s2**2))
    term2 = 2*x - 2*q/(1+q)
    return term1 + term2

def r_pol(r,psi,M1,M2,a):
    q = M2/M1
    x = r*np.cos(psi)
    y = r*np.sin(psi)
    x /= a
    y /= a
    s1 = np.sqrt(x**2 + y**2)
    s2 = np.sqrt((x-1)**2 + y**2)
    
    term1 = 2/(s1*(1+q))
    term2 = 2*q/(s2*(1+q))
    term3 = (x - q/(1+q))**2
    term4 = y**2
    return term1 + term2 + term3 + term4

delta = 0.01
x1 = np.arange(-7, 7, delta)
y1 = np.arange(-7, 7, delta)
X1, Y1 = np.meshgrid(x1, y1)

x2 = np.arange(-17.5,17.5,delta*2.5)
y2 = np.arange(-17.5,17.5,delta*2.5)
X2, Y2 = np.meshgrid(x2, y2)

Z1 = roch2(X1,Y1,1,9.543e-4, 5.2)
Z2 = roch2(X2,Y2,.4,9.543e-4, 13)

ax_len = 14/delta
mid = int(ax_len/2)
t34 = int(3*ax_len/4)

fig, ax = plt.subplots(1,2,figsize=(30,15))
ax[0].plot(Y1[:,t34],-np.log10(abs(Z1[:,t34])), "k", lw=4)
ax[0].plot(Y2[:,t34],-np.log10(abs(Z2[:,t34])), "r", lw=4)
ax[0].set_ylim((-.5,-.47))
ax[0].set_ylabel("log(Potential)", fontsize=24)
ax[0].set_title("Vertical cross section at ~0.5a", fontsize=24)
ax[1].plot(X1[mid,:],-np.log10(abs(Z1[mid,:])), "k", lw=4)
ax[1].plot(X2[mid,:],-np.log10(abs(Z2[mid,:])), "r", lw=4)
ax[1].set_ylim((-.5,-.47))
ax[1].set_title("Cross section through Star and Planet", fontsize=24)
for i in range(2):
    ax[i].hlines([-0.47685,-0.4772,-0.4825,-0.49],-1.4,1.4,"g")
    ax[i].set_yticks([])
    ax[i].set_xticks([])
fig.tight_layout()
fig.show()
```

```python
def r_hill(M1,M2,a):
    return a*((1/3)*(M2/M1))**(1/3)

a1 = 5.2
a2 = 13
M1i = 1
M1f = .4
M2 = 9.543e-4

target1 = 0
target2 = np.pi/3

rlolim1 = a1 - 3*r_hill(M1i, M2, a1)
rhilim1 = a1 + 3*r_hill(M1i, M2, a1)
rlolim2 = a2 - 3*r_hill(M1f, M2, a2)
rhilim2 = a2 + 3*r_hill(M1f, M2, a2)

r1 = np.arange(rlolim1,rhilim1,delta/10)
r2 = np.arange(rlolim2,rhilim2,delta)
psi_L12 = np.arange(-np.pi/6,np.pi/6,delta/10)
psi_L45 = np.arange(np.pi/6,2*np.pi/3,delta/10)

R1, P1 = np.meshgrid(r1,psi_L12)
R2, P2 = np.meshgrid(r2,psi_L12)

R451, P451 = np.meshgrid(r1,psi_L45)
R452, P452 = np.meshgrid(r2,psi_L45)

Zr1 = r_pol(R1,P1,M1i,M2,a1)
Zr2 = r_pol(R2,P2,M1f,M2,a2)
Z41 = r_pol(R451,P451,M1i,M2,a1)
Z42 = r_pol(R452,P452,M1f,M2,a2)

zmax1 = np.amax(Z41)
zmax2 = np.amax(Z42)

zdiff1 = np.log10(abs(zmax1)) - np.log10(abs(Z41))
zdiff2 = np.log10(abs(zmax2)) - np.log10(abs(Z42))

fig, ax = plt.subplots(2,2, figsize=(20,20))

ax[0,0].imshow(np.log10(abs(Zr1)), cmap="Blues", vmin=0.482, vmax=.488, aspect='auto',extent=[rlolim1/a1,rhilim1/a1,target1-np.pi/6,target1+np.pi/6])
ax[0,1].imshow(np.log10(abs(Zr2)), cmap="Blues", vmin=0.482, vmax=.488, aspect='auto',extent=[rlolim1/a1,rhilim1/a1,target1-np.pi/6,target1+np.pi/6])

ax[1,0].imshow(np.log10(abs(Z41)), cmap="Blues", vmin=0.4765, vmax=0.4774, aspect='auto',extent=[rlolim1/a1,rhilim1/a1,target2-2*np.pi/3,target2-np.pi/6])
ax[1,1].imshow(np.log10(abs(Z42)), cmap="Blues", vmin=0.4765, vmax=0.4774, aspect='auto',extent=[rlolim1/a1,rhilim1/a1,target2-2*np.pi/3,target2-np.pi/6])
for n in range(2):
    for i in range(2):
        ax[i,n].set_ylabel(r"$\phi-\phi_{target}$", fontsize=18)
        ax[i,n].set_xlabel(r"$r\,/\,r_p$", fontsize=18)
ax[0,0].set_title("Initial potential", fontsize=18)
ax[0,1].set_title("Potential after mass loss", fontsize=18)
fig.tight_layout()
fig.show()
#fig.savefig("roche_polar.png")
```
