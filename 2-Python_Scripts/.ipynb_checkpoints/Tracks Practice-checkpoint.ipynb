```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline

file = np.loadtxt('2solmass_track.txt')
```

```python
sol_t = file[760:,0]
sol_m = file[760:,1]
sol_l = file[760:,6]

log_l = InterpolatedUnivariateSpline(sol_t, sol_l,k=1)
m_sol = InterpolatedUnivariateSpline(sol_t, sol_m,k=1)
```

```python

step = np.random.randint(0,len(sol_t))
x = np.linspace(sol_t[0],sol_t[-1],1000000)
l = log_l(x)
m = m_sol(x)

plt.scatter(sol_t/1e6,sol_m,s=0.2)
plt.scatter(sol_t/1e6,sol_l,s=0.2)
plt.scatter(x/1e6,l,s=0.001)
plt.plot(x/1e6,m)
#plt.ylim(0.5,4)
plt.xlabel("Age / Myr")
plt.show()
```
