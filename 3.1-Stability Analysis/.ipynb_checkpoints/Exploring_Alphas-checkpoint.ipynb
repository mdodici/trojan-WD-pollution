# Finding $\lambda_i$ for Model A

Working with assumption that when compared to $\frac{27}{4}\mu_2$, $a_0$ is negligibly different for $L_4$ and $L_5$. For now, I've just got this on paper.

```python
import matplotlib.colors as colors
# Example of making your own norm.  Also see matplotlib.colors.
# From Joe Kington: This one gives two different linear ramps:

class MidpointNormalize(colors.Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

# Finding $\alpha_i$

```python
def avg_err(inp,out):
    dis = np.zeros_like(inp)
    err = np.zeros_like(inp)
    for i in range(len(inp)):
        disp = abs(inp[i]-out[i])
        dis[i] = disp
        if inp[i] != 0:
            err[i] = disp/inp[i]
        else:
            err[i] = disp/1e-16
    return sum(dis)/4, dis

def cart_to_pol(x,y):
    r = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y,x)
    return r, phi

def pol_to_cart(r,phi):
    x = r*np.cos(phi)
    y = r*np.sin(phi)
    return x, y
```

```python
A1= np.ones(4)
ks= np.array([0.0782,-0.0782,0.9969,-0.9969])
A2= np.array([4.4059e-8,4.4059e-8,-4.4059e-8,-4.4059e-8])
A3= np.array([-.504,-.504,-1.164,-1.164])
A4= np.array([-2.258e-8,-2.258e-8,-1.098e-7,-1.098e-7])

A = np.array([A1,A2,A3,A4])

P = np.array([-.6,0,.1,0])

a = np.linalg.lstsq(A,P,rcond=None)[0]
b = A3*a
print("a_i =",a)
print('b_i =',b)

P_out = [sum(a),np.dot(a,A2),sum(b),np.dot(a,A4)]
Pout_ = np.dot(A,a)

print("Output values:", Pout_)
print('Average error:',avg_err(P,P_out)[0])
print('Average error:',avg_err(P,Pout_)[0])

print("Start values: X =", P[0],'Y =',P[2])
print("Output values: X =", Pout_[0],'Y =',Pout_[2])

t = np.linspace(0,2e6,100000)
vals = np.ones((4,len(t)))
exps = np.ones((4,len(t)))
coss = np.ones((4,len(t)))
for i in range(len(t)):
    exps[:,i] = np.exp(A2*t[i])
    coss[:,i] = np.cos(ks*t[i])
vals = exps*coss
X = np.ones(len(t))
Y = np.ones(len(t))
for time in range(len(t)):
    X[time] = np.dot(a,vals[:,time])
    Y[time] = np.dot(b,vals[:,time])
    
cx = - max(X[:20000]) + max(X[80000:])
a1 = a[0]; a3 = a[2]
cy = - max(Y[:20000]) + max(Y[80000:])

plt.scatter(t/1e6,np.abs(X),s=0.01)
plt.scatter(t/1e6,np.abs(Y),s=0.01)
plt.show()
```

# Exploring $\alpha_1- \alpha_3$ space

```python
a1test = np.linspace(-.5,.5)
a2test = np.linspace(-.5,.5)

a1_2, a3_4 = np.meshgrid(a1test,a2test)

a_test = np.array([a1_2,a1_2,a3_4,a3_4])

b_test = np.ones_like(a_test)
for i in range(4):
    b_test[i] = A3[i]*a_test[i]
    
t = np.linspace(0,2e6,10000)
vals = np.ones((4,len(t)))
exps = np.ones((4,len(t)))
coss = np.ones((4,len(t)))
for i in range(len(t)):
    exps[:,i] = np.exp(A2*t[i])
    coss[:,i] = np.cos(ks*t[i])
vals = exps*coss
changex = np.zeros((50,50))
changey = np.zeros((50,50))

for i in range(50):
    for n in range(50):
        X = np.ones(len(t))
        Y = np.ones(len(t))
        for time in range(len(t)):
            X[time] = np.dot(a_test[:,i,n],vals[:,time])
            Y[time] = np.dot(b_test[:,i,n],vals[:,time])
        changex[i,n] = -(np.amax(X[:2000]) - np.amax(X[-2000:]))
        changey[i,n] = -(np.amax(Y[:2000]) - np.amax(Y[-2000:]))            
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
dxs = ax[0].pcolor(a1test,a2test,changex,cmap='RdBu_r',norm=MidpointNormalize(midpoint=0))
dys = ax[1].pcolor(a1test,a2test,changey,cmap='RdBu_r',norm=MidpointNormalize(midpoint=0))
for i in range(2):
    ax[i].set_xlabel(r'$\alpha_{\,1,2}$',fontsize=15)
    ax[i].set_ylabel(r'$\alpha_{\,3,4}$',fontsize=15)
    ax[i].set_aspect('equal')
    ax[i].scatter(a1,a3)

ax[0].set_title('Change in x amplitude over 2 Myr')
ax[1].set_title('Change in y amplitude over 2 Myr')

fig.colorbar(dxs,ax=ax[0])
fig.colorbar(dys,ax=ax[1])

fig.tight_layout()
fig.show()
```

# Exploring Physical Space

```python
phi = np.linspace(1.5*np.pi/6,2*np.pi/3,100)
r = np.linspace(0.9,1.1,25)

circp = np.linspace(0,2*np.pi)
circr = np.ones(50)
circx = pol_to_cart(circr,circp)[0] - (0.5 - 0.0009/(1+0.0009))
circy = pol_to_cart(circr,circp)[1] - np.sqrt(3)/2

RR,PH = np.meshgrid(r,phi)
x_i, y_i = pol_to_cart(RR,PH)
x_i -= 0.5 - 0.0009/(1+0.0009)
y_i -= np.sqrt(3)/2

plt.scatter(x_i,y_i,s=0.5)
plt.scatter(0,0,c='k',marker='+')
plt.plot(circx,circy,'k--')
plt.xlim(-1.3,.4)
plt.ylim(-.3,.3)
plt.show()
```

```python
a1 = np.ones_like(x_i)
a3 = np.ones_like(x_i)
for i in range(25):
    for n in range(100):
        P = np.array([x_i[n,i],0,y_i[n,i],0])
        a = np.linalg.lstsq(A,P,rcond=None)[0]
        
        a1[n,i] = a[0]
        a3[n,i] = a[2]

fig, ax = plt.subplots(2,figsize=(10,5))
for i in range(2):
    ax[i].scatter(0,0,c='k',marker='+',zorder=10)
    ax[i].plot(circx,circy,'k--',zorder=8)
    ax[i].set_ylim(-0.25,0.25)
    ax[i].set_xlim(-1.3,0.3)
a___1 = ax[0].pcolor(x_i,y_i,a1,cmap='PuOr',norm=MidpointNormalize(midpoint=0))
a___3 = ax[1].pcolor(x_i,y_i,a3,cmap='PuOr',norm=MidpointNormalize(midpoint=0))
fig.colorbar(a___1, label=r'a1',ax=ax[0])
fig.colorbar(a___3, label=r'a3',ax=ax[1])
fig.show()
```

```python
t = np.linspace(0,2e6,10000)
vals = np.ones((4,len(t)))
exps = np.ones((4,len(t)))
coss = np.ones((4,len(t)))
for i in range(len(t)):
    exps[:,i] = np.exp(A2*t[i])
    coss[:,i] = np.cos(ks*t[i])
vals = exps*coss

a_test = np.array([a1,a1,a3,a3])
b_test = np.ones_like(a_test)
for i in range(4):
    b_test[i] = A3[i]*a_test[i]
    

chnx = np.ones_like(a1)
chny = np.ones_like(a1)
for i in range(25):
    for n in range(100):
        x = np.ones(len(t))
        y = np.ones(len(t))
        for time in range(len(t)):
            x[time] = np.dot(a_test[:,n,i],vals[:,time])
            y[time] = np.dot(b_test[:,n,i],vals[:,time])

        chnx[n,i] = -(np.amax(x[:2000]) - np.amax(x[-2000:]))
        chny[n,i] = -(np.amax(y[:2000]) - np.amax(y[-2000:]))  
```
